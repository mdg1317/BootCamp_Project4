```python
from splinter import Browser
from bs4 import BeautifulSoup
from os.path  import basename
import requests
```

```python
# Set up the browser
browser = Browser('chrome')
```

```python
# Visit the website
url = 'https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number'
browser.visit(url)
```

```python
# Create a BeautifulSoup parser to parse the HTML code
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Scrape all the book information on a single page
all_pokemon = soup.find_all("table", class_="roundy")
all_pokemon
```

```python
# Extract and store the links for each of the books
pokemon_links = []
for gen in all_pokemon:
    links = gen.find_all("td")
    gen_links = [link.find("a") for link in links]
    gen_links = list(filter(None, gen_links))
    gen_hrefs = [link["href"] for link in gen_links]
    gen_hrefs_condensed = [link for link in gen_hrefs if not "_(type)" in link]
    gen_hrefs_condensed = list(set(gen_hrefs_condensed))

    for href in gen_hrefs_condensed:
        pokemon_links.append(href)

pokemon_links
```

```python
pokemon_names = [name[6:-15] for name in pokemon_links]
pokemon_names
```

```python
image_srcs = []
url_start = "https://bulbapedia.bulbagarden.net"

for x in range(len(pokemon_links)):
    url = url_start + pokemon_links[x]

    browser.visit(url)
    soup = BeautifulSoup(browser.html, 'html.parser')

    image = soup.find("img", alt=pokemon_names[x])["src"]
    print(basename(image))

    with open(f"static/images_temp/{basename(image)}", "wb") as f:
        f.write(requests.get(image).content)
```
